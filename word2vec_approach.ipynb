```python
# here we clean the duplicated

df.drop_duplicates(inplace=True)

# check for the duplicated rows

df.duplicated().sum()
```

```python
# here we encode the target values

df["sentiment"]=df["sentiment"].map({"positive":1 , "negative":0 })
df
```

# now we preprocess the data

```python
# import the recuired libraries

import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download("stopwords")
nltk.download("punkt_tab")
```

```python
# tokenize the review column

df["review_token"]=df["review"].apply(word_tokenize)
df
```

```python
# check the tokens size in the first review

len(df["review_token"][0])
```

```python
# convert all tokens in the small form

df["review_token"]=df["review_token"].apply(lambda x:[i.lower() for i in x])
df.head()
```

```python
# now remove the punctuations

df["review_token"]=df["review_token"].apply(lambda x : [i for i in x if i not in string.punctuation])
df.head()
```

```python
# check the tokens size in the first review

len(df["review_token"][0])
```

```python
# now remove the stopwords form the vocabulary

stopword=stopwords.words("english")


df["review_token"]=df["review_token"].apply(lambda x: [i for i in x if i not in stopword])
df.head()
```

```python
# check the tokens size in the first review

len(df["review_token"][0])
```

```python
import gensim
from gensim.models import Word2Vec
```

```python
# apply the word 2 vec model

model=Word2Vec(
    sentences=df["review_token"],
    vector_size=100,
    window=5,
    min_count=2
)
```

```python
model.wv["good"]
```

```python
model.wv.most_similar("production")
```

```python
# convert the each vocabulary to the vector

def vectorizer(vocabulary):

  vector=[]

  for word in vocabulary:
    if word in model.wv:
      vector.append(model.wv[word])

    if len(vector)==0:
      return np.zeros(model.vector_size)

  return np.mean(vector,axis=0)
```

```python
# apply the function on the tokens

df["token_vector"]=df["review_token"].apply(vectorizer)
df.head()
```

```python
# now we create the independant and dependant variables

x=df["token_vector"].tolist()
y=df["sentiment"].tolist()
```

```python
# now split into the train and test sets

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

- here we preprossed the data from convert the vocabulary into the vectors

# Now we train the machine learning model

```python
# first we train on the logistic regression

from sklearn.linear_model import LogisticRegression

linear_model=LogisticRegression()
linear_model.fit(x_train,y_train)
```

```python
# now evaluate the models performance on test set

from sklearn.metrics import accuracy_score,classification_report

y_pred=linear_model.predict(x_test)

print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

```python
# now evaluate the models performance on train set

from sklearn.metrics import accuracy_score,classification_report

y_pred=linear_model.predict(x_train)

print(accuracy_score(y_train,y_pred))
print(classification_report(y_train,y_pred))
```

```python
# here we try the hyperparameter tuning

from sklearn.model_selection import GridSearchCV

logi_model=LogisticRegression()

param_grid={"C":[0.01,0.1,1.0],
            "solver":["liblinear","saga"],
            "penalty":["l1","l2"]}


grid=GridSearchCV(logi_model,param_grid,cv=5,scoring="accuracy")
grid.fit(x_train,y_train)
```

```python
# train and evalueate with best estamators

logi_model=LogisticRegression(**grid.best_params_)

logi_model.fit(x_train,y_train)

y_pred=logi_model.predict(x_test)

print(accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
```

- there is no such a big diffrence between the hyperparameter and the custom parameters
