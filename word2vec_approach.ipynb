```python
# import the dataset

df=pd.read_csv("IMDB-Dataset.csv")
df.head()
```

```python
# check for the summary of the data

df.info()
```

- we have two columns first is the review which contain the review about the movie.

- and in sentiment column we have the two values positive and negative values .

```python
# check for the null values

df.isna().sum()
```

- there is no any null values in the whole dataset.

```python
# check for the duplicates

df.duplicated().sum()
```

- there is an 418 duplicated rows in the dataset.

```python
# here we clean the duplicated

df.drop_duplicates(inplace=True)

# check for the duplicated rows

df.duplicated().sum()
```

```python
# here we encode the target values

df["sentiment"]=df["sentiment"].map({"positive":1 , "negative":0 })
df
```

# now we preprocess the data

```python
# import the recuired libraries

import nltk
import string
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download("stopwords")
nltk.download("punkt_tab")
```

```python
# tokenize the review column

df["review_token"]=df["review"].apply(word_tokenize)
df
```

```python
# check the tokens size in the first review

len(df["review_token"][0])
```

```python
# convert all tokens in the small form

df["review_token"]=df["review_token"].apply(lambda x:[i.lower() for i in x])
df.head()
```

```python
# now remove the punctuations

df["review_token"]=df["review_token"].apply(lambda x : [i for i in x if i not in string.punctuation])
df.head()
```

```python
# check the tokens size in the first review

len(df["review_token"][0])
```

```python
# now remove the stopwords form the vocabulary

stopword=stopwords.words("english")


df["review_token"]=df["review_token"].apply(lambda x: [i for i in x if i not in stopword])
df.head()
```

```python
# check the tokens size in the first review

len(df["review_token"][0])
```

```python
import gensim
from gensim.models import Word2Vec
```

```python
# apply the word 2 vec model

model=Word2Vec(
    sentences=df["review_token"],
    vector_size=100,
    window=5,
    min_count=2
)
```
